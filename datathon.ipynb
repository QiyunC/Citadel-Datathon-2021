```python
import matplotlib.pyplot as plt
import geopandas as gpd

countries = gpd.read_file(  
     gpd.datasets.get_path("naturalearth_lowres"))
# countries[countries["name"] == "Australia"].plot(color="lightgrey",
#                                                  ax=ax)
countries['centroid'] = countries['geometry'].to_crs('+proj=cea').centroid.to_crs(epsg=4326)
countries['centroid_x'] = countries['centroid'].x
countries['centroid_y'] = countries['centroid'].y

uniq_c_set = set(countries['name'].unique())
i = 0
not_included_lst = []
for c in tobacco_use_ww_df['Location'].unique():
    if c in uniq_c_set:
        i += 1
    else:
        not_included_lst.append(c)

tobacco_use_ww_df['Location'] = tobacco_use_ww_df['Location'].map(lambda x: d[x] if x in d else x)
# merge
total_df = pd.merge(
    countries, 
    tobacco_use_ww_df[tobacco_use_ww_df['Gender'] == 'Both sexes'],
#         (tobacco_use_ww_df['Gender'] == 'Both sexes') & \
#         (tobacco_use_ww_df['Year'] == 2018)],
    left_on="name", 
    right_on="Location")
# total_df['Year'] = pd.to_datetime(total_df['Year'], format='%Y')
total_df.sort_values('Year', ascending=True, ignore_index=True, inplace=True)
```

```python
total_df['Year'] 
```

```python
range_color = [total_df['Value'].min(), total_df['Value'].max()]
range_year = [total_df['Year'].min(), total_df['Year'].max()]
```

```python
def plot1(year):
    fig, ax = plt.subplots(figsize=(15, 15))
    total_df_yr = total_df[total_df['Year'] == year]
    countries.plot(color='lightgrey', ax=ax)
    # plot points
    total_df_yr.plot(x="centroid_x", y="centroid_y", kind="scatter", 
        c="Value", colormap="YlOrRd",  
        title="Worldwide Tobacco Usage in %d" % year, 
        ax=ax)
    # add grid
    ax.grid(b=True, alpha=0.5)
    plt.show()
```

```python
import plotly.express as px
# color by parent location 
def plot2():
    total_df_yr = total_df
    fig = px.scatter_geo(total_df_yr, lat='centroid_y', lon='centroid_x',
                     size='Value', size_max =12, color='ParentLocation', # size of markers, "pop" is one of the columns of gapminder
                     color_discrete_sequence=px.colors.qualitative.Set3,
                     hover_data=['Location','ParentLocation'],
                     animation_frame="Year",
                     title="Worldwide Tobacco Usage between %d and %d" % 
                         (range_year[0], range_year[1]))
    fig.show()
```

```python
plot2()
```

```python
def plot3():
    total_df_yr = total_df
    fig = px.scatter_geo(total_df_yr, lat='centroid_y', lon='centroid_x',
                     size='Value', size_max =12, color='Value', # size of markers, "pop" is one of the columns of gapminder
                     color_continuous_scale=px.colors.diverging.Temps,
                     range_color=range_color,
                     hover_data=['Location','ParentLocation'],
                     animation_frame="Year",
                     title="Worldwide Tobacco Usage between %d and %d" % 
                         (range_year[0], range_year[1]))
    fig.show()
    
```
